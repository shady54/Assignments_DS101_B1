```
#10. Plot distribution of match's duration vs walk distance one below another.

import matplotlib.pyplot as plt

plt.style.use('ggplot')
plt.figure()
plt.subplot(2,1,1)
plt.plot(pubg.matchDuration,'-',);
plt.xlabel("MatchDuration")

plt.subplot(2,1,2)
plt.plot(pubg.walkDistance,'--');
plt.xlabel("WalkDistance")
```

```
#11. Plot distribution of match's duration vs walk duration side by side.

import matplotlib.pyplot as plt

plt.style.use('ggplot')
plt.figure()
plt.subplot(1,2,1)
plt.plot(pubg.matchDuration,'-',);
plt.xlabel("MatchDuration")

plt.subplot(1,2,2)
plt.plot(pubg.walkDistance,'--');
plt.xlabel("WalkDistance")
```

```
#22. Take a sample of size 50 from the column damageDealt for 100 times and calculate its mean.
#Plot it on a histogram and comment on its distribution.
import seaborn as sns
mean = []

for i in range(100):
  
  for x in range(0,1001,50):
    
    mean1 = pubg['damageDealt'].head(x).mean()
    mean.append(mean1)
 
sns.histplot(data=pubg['damageDealt'], x=mean);
```
